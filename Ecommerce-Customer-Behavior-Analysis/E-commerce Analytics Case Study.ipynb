```python
import pandas as pd
import numpy as np

df = pd.read_csv("ecommerce_customer_behavior_dataset_v2.csv")

df.head()
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df[df['Quantity'] <= 0]
```

```python
df['Calculated_Total'] = df['Unit_Price'] * df['Quantity'] - df['Discount_Amount']
```

```python
df.drop(columns="Calculated_Total", inplace=True)
```

```python
df['Product_Category'] = df['Product_Category'].str.strip()
df['Gender'] = df['Gender'].str.title()
```

```python
df['Discount_Percentage'] = round((df['Discount_Amount'] / (df['Unit_Price'] * df['Quantity'])) * 100,2)
```

```python
bins = [18,25,35,45,55,65,80]
labels = ['18-24','25-34','35-44','45-54','55-64','65+']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df.to_csv("Clean_E-Commerce_Data.csv", index=False)
```
